```python
def calculate_bmi(weight, height):
    bmi= weight/(height**2)
    if bmi < 18.5:
        return "underweight"
    elif 18.5 < bmi < 24.9:
        return "Normal"
    elif (bmi<=29.9) and (bmi>=25):
        return "overweight"
    elif 29.9 < bmi :
        return "very overweight"
```

```python
df["weight_string"]=df["weight"].apply(str)
df["weight_duplicated"]=df["weight"].apply(lambda x:x*2)
df
```

```python
df["bmi"]=df.apply(lambda row: calculate_bmi(row["weight"], row["height"]),axis=1)
df[["weight", "height", "bmi"]]
```
